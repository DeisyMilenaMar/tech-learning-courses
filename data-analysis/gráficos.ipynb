### Correlación estadística

- Mide la relación entre dos variables,  y cómo se influencian mutuamente.
* **Ejemplo 1**: Fumar y el cáncer de pulmón están correlacionados: fumar aumenta el riesgo de cáncer.
* **Importante**: La correlación no implica causalidad. Por ejemplo, paraguas y lluvia están correlacionados, pero los paraguas no causan lluvia.

#### Correlación de Pearson

- Correlación de Pearson NO es adecuado para calcular la matriz de correlación de todo un DataFrame de una sola vez.
- **pearsonr** está diseñado para calcular la correlación entre dos series de datos individuales (x e y), no entre múltiples columnas al mismo tiempo.

##### a). **Coeficiente de correlación**

- Valores cercanos a 1 → *Fuerte correlación positiva*.
- Valores cercanos a -1 → *Fuerte correlación negativa*.
- Valores cercanos a 0 → *Sin correlación*.

##### b). **Valor p**

- Mide la certeza estadística de la correlación.
- **p < 0.001:**   *Alta certeza*.
- **0.001 ≤ p < 0.05:**   *Certeza moderada*.
- **0.05 ≤ p < 0.1:**   *Baja certeza*.
- **p > 0.1:**   *Sin certeza*.

##### d). **Calcular coeficiente de correlación y p_value** 
para la columna 'Energy_kcal' & 'Protein_g'

```python
from scipy.stats import pearsonr

# Ejemplo: correlación entre Energy_kcal y Protein_g
x = df['Energy_kcal']
y = df['Protein_g']

# Calcular coeficiente de correlación y valor p
pearson_coef, p_value = pearsonr(x, y)

print(f"Pearson Coefficient: {pearson_coef}")
print(f"P-value: {p_value}")
```

##### e). **Calcular coeficiente de correlación y p_value** 
para las columnas 'Energy_kcal', 'Protein_g', 'Fat_g', 'Carb_g'

```python
from scipy.stats import pearsonr
import pandas as pd
import numpy as np

# Seleccionar columnas específicas
columns_to_include = ['Energy_kcal', 'Protein_g', 'Fat_g', 'Carb_g']
selected_data = df[columns_to_include]

# Crear una matriz vacía para almacenar los coeficientes de Pearson
correlation_matrix = pd.DataFrame(
    np.zeros((len(columns_to_include), len(columns_to_include))),
    index=columns_to_include,
    columns=columns_to_include
)

# Llenar la matriz con los coeficientes de Pearson
for col1 in columns_to_include:
    for col2 in columns_to_include:
        if col1 == col2:
            correlation_matrix.loc[col1, col2] = 1.0  # Correlación perfecta consigo misma
        else:
            pearson_coef, _ = pearsonr(selected_data[col1], selected_data[col2])
            correlation_matrix.loc[col1, col2] = pearson_coef

print(correlation_matrix)
```

#### e). Método pandas.corr()

- El método pandas.corr() es más eficiente que usando Pearson para calcular la matriz de correlación de todo el DataFrame de una vez

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Seleccionar columnas específicas
columns_to_include = ['Energy_kcal', 'Protein_g', 'Fat_g', 'Carb_g']
selected_data = df[columns_to_include]

# Calcular la matriz de correlación
correlation_matrix = selected_data.corr()

# Crear el heatmap
plt.figure(figsize=(8, 6))  # Ajusta el tamaño del gráfico
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Configurar títulos y etiquetas
plt.title("Heatmap de Correlaciones (Columnas Seleccionadas)", fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
